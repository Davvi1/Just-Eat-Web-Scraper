```python
def get_menu_link(location, url):
    """Function for finding unique part of url.
    
    Function opens webdriver, clicks cookies, searches for location,
    searches for pizza, saves html, finds the unique part of the url
    for some pizzeria for each location.
    
    Args:
        location (list of (str)): list of strings with locations
        url (str): url of website
        
    Returns:
        menu_links(list of (str)): unique part of pizzeria urls
    """
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--incognito")
    driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
    driver.implicitly_wait(100)
    driver.get(url)
    driver.maximize_window()
    
    time.sleep(2)
    cookie = driver.find_element(By.CSS_SELECTOR, '._2JFg2') 
    cookie.click()
    
    time.sleep(5)
    search_location = driver.find_element(By.CSS_SELECTOR, '#combobox-input_0')
    search_location.send_keys(location)
    search_location.send_keys(Keys.RETURN)
    
    time.sleep(5)
    search_location.send_keys(Keys.DOWN, Keys.ENTER, Keys.ENTER)
    
    time.sleep(5)
    search_food = driver.find_element(By.CSS_SELECTOR, '._3V5BI')
    search_food.send_keys('pizza')
    search_food.send_keys(Keys.RETURN)
    
    time.sleep(5)
    html = driver.page_source
    soup = BeautifulSoup(html)
    find_links = soup.find_all('a')
   
    time.sleep(2)
    menu_links = []

    for link in find_links:
        if link['href'][1:5] == 'menu':
            menu_links.append(link['href'])     
    driver.close()
    return menu_links
```

```python
#Extracting HTML
wiki_url = 'https://da.wikipedia.org/wiki/Danmarks_st%C3%B8rste_byer'
response = requests.get(wiki_url)
soup = BeautifulSoup(response.content, 'html.parser')
tables = soup.find_all('table', class_ = 'wikitable sortable')
poptable = tables[1]
```

```python
#Extracting rows and values
rows = poptable.find_all('tr')

result = []
for row_node in rows:
    row = []
    for child in row_node.children:
        row.append(child.text)
    result.append(row)
```

```python
#Creating Dataframe...
messy_table = pd.DataFrame(result)
columned_table = messy_table.rename(columns=messy_table.iloc[0]).loc[1:]
dropped_table = columned_table.drop(['\n', '#', '+/-[b]\n', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'], axis = 1)
juicy_table = dropped_table.replace(regex = [r'\n+'], value = '')
juicy_table = juicy_table.replace('\[\w\]', '', regex=True)
juicy_table = juicy_table.replace(r'\([^()]*\)', '', regex=True)
juicy_table.rename(columns={'Byområde\n': 'Byområde', '2022' : 'Population 2022'}, inplace=True)
juicy_table.at[1, 'Byområde'] = 'København K'
juicy_table.at[3, 'Byområde'] = 'Odense C'
juicy_table.at[14, 'Byområde'] = 'Helsingør by'
juicy_table.at[33, 'Byområde'] = 'Farum by'
```

```python
links = []
for location in denmark:
    menu_link_function = get_menu_link(location, url)
    links.append(menu_link_function)
```

```python
res_names = [] # Create new list (of lists) for inputting restaurant names
for city_links in links: # For each bundle of links per city...
    res_names_from_city = []
    for link in city_links: # For the individual restaurant link in the bundle of city links
        name_of_res = re.findall(r'/menu/(\S+)', link) # Find the restaurant name
        res_names_from_city.append(name_of_res)
    flat_list = [item for sublist in res_names_from_city for item in sublist]
    res_names.append(flat_list) # Add this new bundle of names to the list of list of names
```

```python
list_of_dict = dict(zip(denmark, res_names))
```

```python
place_links = [] # Insert the links to the URL.

for link in range(len(links)):
    list_in_list = links[link]
    city_links = []
    
    for i in list_in_list:
        url_place = f'https://www.just-eat.dk{i}' # Some URL goes to the frontpage (eg. diegos), need to be removed with 'try except'
        city_links.append(url_place)
    place_links.append(city_links)
```

```python
def pizza_type_price(pizza_type):
    """Function for finding pizza prices.
    
    Function opens webdriver for each pizzaria link, clicks cookies, 
    searches for specific pizza type, saves html, use html to find
    pizza price and appends to list. Appends NaN if no price is found.
    
    Args:
        pizza_type (str): some pizza type

    Returns:
        price_list(list of (str)): list of prices for each pizza type, 
        each pizzeria, each location
    """
    
    price_list = []

    for i in range(len(place_links)):
        use_links = place_links[i]
        for j in range(len(use_links)):
            chrome_options = webdriver.ChromeOptions()
            chrome_options.add_argument("--incognito")
            place_driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
            place_driver.implicitly_wait(100)
            place_driver.get(use_links[j]) # Use the url to the page for the different restaurants.
            place_driver.maximize_window()

            place_cookie = place_driver.find_element(By.CSS_SELECTOR, '._2JFg2') # Find cookie
            place_cookie.click()
            time.sleep(2)

            search1 = place_driver.find_element(By.CSS_SELECTOR, "div[aria-label='Søger efter produkt']").click() # Click on search icon
            #search1.clear()

            xpath = "//input[contains(@data-qa, 'menu-category-nav-search-element')]"
            search_bar = place_driver.find_element(By.XPATH, xpath)  # Find searchbar
            search_bar.send_keys(pizza_type) 
            time.sleep(2)

            html_place = place_driver.page_source # Get the current HTML
            soup_place = BeautifulSoup(html_place) # Make the HTML to BeutifulSoup(BS)
            pizza_menu = soup_place.find_all('div', class_ = '_1sMed _1wlHd Yc8ZH _1VvW7 hzg-6')

            items = [] # List for links to restaurants
            for i in range(len(pizza_menu)):
                if pizza_menu[i].find('h3').text.__contains__(pizza_type): # The first 1-5 needs to be menu in the href to get the correct links.
                    items.append(pizza_menu[i]) # Append links to a list
            try:
                items = items[0].find(class_ = '_2PRj3E').text
                item_price = re.sub('[^0-9,]','',items)
                price_list.append(item_price)
            except IndexError:
                title = 'IndexError'
                price_list.append(np.NaN)
    
    return price_list
```
